## Checking Empty Files

##SEP28K

```python
import os

# Specify the root directory where the clips are stored
root_dir = '/content/drive/MyDrive/Clips'

# Function to get a dictionary of files and their sizes
def get_files_and_sizes(root_dir):
    file_sizes = {}
    empty_files = []
    for root, dirs, files in os.walk(root_dir):
        for file in files:
            clip_path = os.path.join(root, file)
            file_size = os.path.getsize(clip_path)
            file_sizes[clip_path] = file_size
            if file_size <= 44:
                empty_files.append(clip_path)
    return file_sizes, empty_files

# Call the function to get a dictionary of files and their sizes, and a list of empty files
file_sizes, empty_files = get_files_and_sizes(root_dir)

# Print the dictionary of files and their sizes
for file, size in file_sizes.items():
    print(f"File: {file}, Size: {size} bytes")

# Print the list of empty files
if empty_files:
    print("Empty files:")
    for file in empty_files:
        print(file)
else:
    print("No empty files found.")
```

# Feature Extraction

##Sep-28k_feature's

```python
import librosa
import numpy as np
from tqdm.notebook import tqdm
import os
import pickle

# Define the root directory where the clips are stored
root_dir = '/content/drive/MyDrive/Clips'

# Define the main directories to process
main_directories = ['HVSA', 'HeStutters', 'IStutterSoWhat', 'MyStutteringLife', 'StutterTalk', 'WomenWhoStutter']

sep28k_features = {}

# Check if features have already been extracted and saved
if os.path.exists('/content/drive/MyDrive/sep28k_features.pkl'):
    print("Features have already been extracted and saved.")
    with open('/content/drive/MyDrive/sep28k_features.pkl', 'rb') as file:
        sep28k_features = pickle.load(file)
else:
    # Define a function to extract MFCC features
    def extract_mfcc_features(clip_path):
        audio, sample_rate = librosa.load(clip_path, res_type='kaiser_fast', sr=None)
        mfccs = np.mean(librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=13).T, axis=0)
        return mfccs

    for main_directory in main_directories:
        main_directory_path = os.path.join(root_dir, main_directory)

        for root, dirs, files in os.walk(main_directory_path):
            for filename in tqdm(files):
                if filename.endswith('.wav'):
                    clip_path = os.path.join(root, filename)
                    clip_name = filename[:-4]  # Remove the ".wav" extension

                    # Check if the clip should be ignored (no ignore list used here)
                    if 'FluencyBank' not in clip_name:
                        mfccs = extract_mfcc_features(clip_path)
                        sep28k_features[clip_name] = mfccs


    # Save the extracted features
    with open('/content/drive/MyDrive/sep28k_features.pkl', 'wb') as file:
        pickle.dump(sep28k_features, file)
    print("Features have been extracted and saved.")


# Print a sample feature from sep28k_features
sample_clip_name = list(sep28k_features.keys())[0]
sample_feature = sep28k_features[sample_clip_name]

print(f"Sample feature for clip '{sample_clip_name}':")
print(sample_feature)
```

##FluencyBank

```python
import librosa
import numpy as np
from tqdm.notebook import tqdm
import os

# Define the root directory where the clips are stored
root_dir = '/content/drive/MyDrive/fluencybank_clips'

# Define a function to extract MFCC features
def extract_mfcc_features(clip_path):
    audio, sample_rate = librosa.load(clip_path, res_type='kaiser_fast', sr=None)
    mfccs = np.mean(librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=13).T, axis=0)
    return mfccs

# Create an empty dictionary to store the extracted features
fluencybank_features = {}

# Define the directory for the FluencyBank dataset
fluencybank_directory = os.path.join(root_dir, 'FluencyBank')

# Loop through the directory and its subdirectories, extracting MFCC features
for root, dirs, files in os.walk(fluencybank_directory):
    for filename in tqdm(files):
        if filename.endswith('.wav'):
            clip_path = os.path.join(root, filename)
            clip_name = filename[:-4]  # Remove the ".wav" extension

            mfccs = extract_mfcc_features(clip_path)
            fluencybank_features[clip_name] = mfccs

# Print a sample feature for verification
sample_clip_name = list(fluencybank_features.keys())[0]
sample_feature = fluencybank_features[sample_clip_name]
print(f"Sample feature for clip '{sample_clip_name}':")
print(sample_feature)
```
